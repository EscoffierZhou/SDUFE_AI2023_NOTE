## **2.1   程序设计基本方法**
>**I  input**
>
>>**文件输入、网络输入、控制台输入、交互界面输入、随机数据输入、内部参数输入**
>
>**P  process**
>
>>**处理数据**
>
>**O output**
>
>>**输出的方式有很多种：文件输出、网络输出、控制台输出、内部参数输出**


````python
#This is a comment by 202318140413
#Example：Temperature convert (Straightly)
TempStr=input("Insert temperature with symbol")
if(TempStr[-1] in ['F']or['f']):  ##注意格式in和方括号[]
#The same: if TempStr[-1] in ('F', 'f'):
    C=(eval(TempStr[0:-1])-32)/1.8
    #//The same        C=(eval(TempStr[(0):(-1)])-32)/1.8 
    print("The converted temperature is %.2f C" % (C))
    #//The same        print("The converted temperatrue is {:.2f}C".format(C))
elif TempStr[-1] in['C']or['c']:
    F=1.8*eval(TempStr[0:-1])+32
    print("The converted temperatrue is {:.2f}F".format(F))
    #//The same        print(f"The converted temperature is{F:.2f}")
else:
    print("ERROR")
````

## **2.2  程序基本语法元素**

**缩进/注释/命名和保留字/字符串/赋值语句/input()/分支语句/eval()/print()/循环语句/定义函数**
>**1 缩进   $\quad \quad$一行代码开始前的空白区域，表达代码之间的包含和层次关系**
>
>>**[if-elif-else][for while] 都需要用缩进表达对应关系。缩进可以包括多层**
>
>**2 注释    $\quad \quad$用于提高代码可读性的辅助性文字，不被执行**
>
>>**单行注释：以#开头，其后内容为注释**
>>
>>**多行注释：以成对的'''开头和结尾或"""**
>>
>**3 变量    $\quad \quad$程序中用于保存和表示数据的占位符号，是标识符的一种。**
>
>>**注意事项: 大小写敏感、首字符不能是数字、不与保留字相同**
>>
>**4 保留字  $\quad \quad$被编程语言内部定义并保留使用的标识符**
>
>>
>>
>**5 数据类型 $\quad \quad$字符串、数字、列表**
>
>>**字符串是由若干个（0个或多个)字符组成的有序字符序列**
>>
>>**列表是由[]括起来的若干个（0个或多个)元素组成的有序序列**
>>
>**6  赋值语句 $\quad \quad$由赋值符号构成的一行代码**
>
>**7  分支语句 $\quad \quad$由判断条件决定程序运行方向的语句**
>
>**8  函数    $\quad \quad$根据函数不同和输入参数产生不同输出的过程，如 input( )  eval( )  print( )**
