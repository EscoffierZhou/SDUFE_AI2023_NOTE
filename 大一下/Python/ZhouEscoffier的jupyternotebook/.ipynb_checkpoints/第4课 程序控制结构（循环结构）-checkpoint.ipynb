## **第4课 程序控制结构（循环结构）**

1. **程序有三种基本结构组成：顺序结构，循环结构，分支结构**
2. **顺序结构是指程序按照顺序依次进行的运行方式**
3. **分支结构是指（for循环或者while循环）程序按照双分支或多分支的形式运行（注意只能n选1）`for i in range(start,stop,sequence)`**
4. **循环结构是指（while循环）程序按照循环次数运行程序，其中有关键词continue跳过当前循环，break关键词跳出所有循环**

**程序的循环结构（都有用！！）**
>**遍历循环  for （常用于确定次数的循环)**
>
>**无限循环  while （常用于不确定次数的循环)**
>
>**循环控制保留字 break continue**
>
>**循环的高级用法 与else的配合使用**
>

**While循环：常用于无限循环**
>**由while+条件控制的循环运行方式,反复执行语句块，直到条件不满足时结束.**
>
>>**但是有一种没有终止条件的循环：‘死循环’一般是无益的，但是还是有用的**
>>
>>**实际上有些自动设备依赖死循环（例如：实时监控设备、浏览器打开后自动刷新等）**
>>
>>>**示例：**

>**循环的控制保留字 break 和 continue循环后的else语句**
>
>>**break 和 continue**
>>
>>>**break跳出并结束当前层循环（教材上：最内层循环），执行循环后的语句**
>>>
>>>>```python
>>>># break 示例
for s in 'BIT': #循环变量s遍历字符串‘BIT’
    for i in range(4): #原本内层会循环4次
        print(s,end='')
        if s == 'I':
            break #‘I'先输出1次，然后退出当前层循环
>>>
>>>**continue不再执行当前层循环剩余的语句，继续执行下次循环**
>>>```python
>>># continue 示例
for s in 'BIT': #循环变量s遍历字符串‘BIT’
    for i in range(4): #原本内层会循环4次
        print(s,end='')
        if s == 'I':
            continue # 看起来continue没有起作用
>>>
>**循环的高级用法，循环和else语句配对使用**
>>
>>>**在循环中执行过break语句，就不执行else语句；其他情况都执行else语句**
>>>
>>>**else语句块作为"正常"完成循环的奖励continue不会对else执行造成影响**

#### **圆周率计算：**
**计算机中圆周率不能直接被算出来，在math库里面的pi是近似公式但不是泰勒展开算出来的**
>>**所以不能用于高精度计算**
>>
**$$\pi \quad = \quad \ \sum_{k=0}^{\infty} \ [\frac{1}{16^k}(\frac{1}{8k+1}-\frac{2}{8k-4}-\frac{1}{8k+5}-\frac{1}{8k+6})]$$**
```python
pi=0
for k in range(0,20):
    pi += (1.0/(16**k))*(4.0/(8*k+1)-2/(8*k+4)-1/(8*k+5)-1/(8*k+6))
print(pi)
# 3.141592653589793
```
**蒙特卡罗方法：随机点方法**
>**使用随机函数落点在1*1的正方形，再写出圆形轨迹**
>>>>>
>
>```python
>#CalPiV2.py
from random import random  DARTS = 1000*1000
hits = 0.0
for i in range(1, DARTS+1):  x, y = random(), random()
dist = pow(x ** 2 + y ** 2, 0.5)
if dist <= 1.0:  hits = hits + 1
pi = 4 * (hits/DARTS)
print("圆周率值是: {}".format(pi))
>```

````python
"""
6. (简答题, 10分)
编写代码实现计算累加和功能。

要求：用户输入一个正整数N，计算机能自动计算出从0到N的整数之和。
"""
#This is a comment by 周兴 202318140413

sum = 0
for i in range(101):
    sum = sum + i
print(sum)

sum = 0
i = 0
while i < 101:
    sum = sum + i
    i += 1
print(sum)
````

````python
"""
编写代码实现猜数字密码的功能。

要求：计算机先随机生成一个两位数的密码【密码的范围是[1,99]】，并将该密码输出；
用户通过键盘输入所猜之数。
如果此数大于预设的密码，则显示提示“遗憾，太大了”； 
如果此数小于预设的密码，则显示提示“遗憾，太小了”；如果此数等于预设的密码，则显示提示“恭喜，你猜对了”
"""
#This is a comment by 周兴 202318140413
import random

x = random.randint(1, 99)

while 1:
    i = eval(input("请输入1~99之间的一个数："))
    if i == x:
        print("你真是个小天才！")
        break
    elif i < x:
        print("小混蛋，猜小了！")
    else:
        print("小混蛋，猜大了！")
````

````python
"""
编写代码实现输出3的倍数。

要求：将100以内的3的倍数，按照从大到小的顺序，输出最大的5个值。
"""
#This is a comment by 周兴 202318140413
x=100
cnt =0
while 1:
    if x%3==0:
        print("%d" % x)
        cnt+=1
    x-=1
    if cnt==5:
        break
````
