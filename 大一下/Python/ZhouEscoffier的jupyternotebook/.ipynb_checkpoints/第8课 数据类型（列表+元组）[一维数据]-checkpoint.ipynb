````python
#This is a comment by  202118140413
numbers = []
while True: 
    try:
        num = float(input(""))  
    except ValueError:  
        break
    else: 
        numbers.append(num)

average = sum(numbers)/len(numbers) 
numbers.sort()
index = int((len(numbers))/2) 
if len(numbers) % 2 == 0:  
    mid = (numbers[index-1] + numbers[index])/2
else:
    mid = numbers[index]

print("{:.1f}".format(average))  
print("{:.1f}".format(mid))
````
