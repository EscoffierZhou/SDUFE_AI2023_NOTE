#### **4.1.4 元组类型（tuple）**
**元组类型及操作(简单)**
>**定义:元组也是一种序列类型,元组独特之处：一旦创建就不能修改，只能访问**
>
>**创建：用( )或tup( )创建，甚至可以不用( )**
>
>**操作：元组可以用 +  $*$  in操作**
>
>**使用：函数的返回值就是元组类型.[return 1,2]**
>
>>>```python
>>> creature= tuple("cat", "dog", "tiger", "human")
>>> creature[::-1]
('human', 'tiger',	'dog', 'cat')
>>> color = (0x001100, "blue", creature)
>>> color[-1][2]  'tiger'
>>>```
>
**序列类型应用场景**
>**列表是存储和检索的有序序列，通过整数的索引来查找目标元素，但是实际上需要更加有效的检索方式**
>
>>**通过任意键信息查找一组数据中元素信息的过程叫做映射，Python里面叫做字典，通过 `{}`进行表示**
>
>**列表更加灵活，它是最常用的序列类型**
>
>**最主要作用：表示一组有序数据，进而操作它们**

````python
import random
guess = []
for i in range(3):
    n = random.randint(0, 9)
    guess.append(n)
print(guess)

for i in range(len(guess)):
    if guess.count(i) > 1:
        print("列表中有重复元素")
        break
````

**17. (简答题, 20分)基本统计值计算。用户输入若干个数生成列表，若输入的非数字则停止接收输入；计算以输入这些数的平均值和中位数，并输出。并用[1,2,3,4,5] 和 [1,2,3,4,5,6] 验证结果的正确性。示例：**


````python
#This is a comment by 周兴 202118140413
numbers = []
while True: 
    try:
        num = float(input("请输入一个数："))  
    except ValueError:  
        break
    else: 
        numbers.append(num)

average = sum(numbers)/len(numbers) 
numbers.sort()
index = int((len(numbers))/2) 
if len(numbers) % 2 == 0:  
    mid = (numbers[index-1] + numbers[index])/2
else:
    mid = numbers[index]

print("平均值是{:.1f}".format(average))  
print("中位数是{:.1f}".format(mid))
````
