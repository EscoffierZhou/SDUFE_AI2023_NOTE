````python
#显示随机数关系的散点图.py

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 4 + np.random.normal(0,2,240)
y = 4 + np.random.normal(0,2,len(x))

#size and color
sizes = np.random.uniform(15,500,len(x)) #通过张量才能设置单独一个的大小
colors = np.random.uniform(-1000,1000,len(x))

#plot
plt.scatter(x,y,s=sizes,c=colors,vmin=0,vmax=1000)
plt.show()
````

````python
import matplotlib.pyplot as plt  
import numpy as np  

plt.figure(figsize = (6,5))


x = 0.5 + np.arange(6)  # x的值从0.5开始，到5.5结束，共6个值  
y = np.random.randint(0, 100, size=len(x))  # 生成与x长度相同的随机整数数组，范围从0到99  
y2 = np.random.randint(0, 100, size=len(x))  # 生成与x长度相同的随机整数数组，范围从0到99
y3 = np.random.randint(0, 100, size=len(x))  # 生成与x长度相同的随机整数数组，范围从0到99
plt.subplot(5,1,1)
plt.bar(x, y, width=0.8, color="pink", linewidth=0)  
plt.show()
plt.subplot(5,1,2)
plt.hist(x)  
plt.show()
plt.subplot(5,1,3)
plt.barh(x, y2,height = 0.5, color="purple", linewidth=0)  # 如果是barh则是横向图
plt.show()
plt.subplot(5,1,4)
colorMap = plt.set_cmap('jet')
explodes = [0.3,0.1,0,0,0,0] #这个是序列对应的偏离数值，不是索引
plt.pie(y3,explode = explodes,labels = x, colors=colorMap)  # 如果是barh则是横向图
plt.show()
````

````python
# 绘制“霍兰德人格分析雷达图”
import numpy as np      # 专业的多维数据表示
import matplotlib.pyplot as plt
import matplotlib      # 通用雷达图绘制
matplotlib.rcParams['font.family']='SimHei'                    # 为了正确显示中文字体，将字体更改为黑体‘SimHei’
radar_labels = np.array(['研究型(I)','艺术型(A)','社会型(S)',\
                         '企业型(E)','常规型(C)','现实型(R)'])  # 雷达标签
nAttr = 6
data = np.array([[0.40, 0.32, 0.35, 0.30, 0.30, 0.88],
                 [0.85, 0.35, 0.30, 0.40, 0.40, 0.30],
                 [0.43, 0.89, 0.30, 0.28, 0.22, 0.30],
                 [0.30, 0.25, 0.48, 0.85, 0.45, 0.40],
                 [0.20, 0.38, 0.87, 0.45, 0.32, 0.28],
                 [0.34, 0.31, 0.38, 0.40, 0.92, 0.28]])      # 数据值
data_labels = ('艺术家', '实验员', '工程师', '推销员', '社会工作者','记事员')
angles = np.linspace(0, 2*np.pi, nAttr, endpoint=False)     # 将360度平均分为n个部分（有endpoint=False分为6个部分，反之5个部分）
 
plt.figure(facecolor="white")   # 绘制全局绘图区域
plt.subplot(111, polar=True)    # 绘制一个1行1列极坐标系子图，当前位置为1
 
plt.plot(angles,data,'o-', linewidth=1, alpha=0.2)     # 连线，画出不规则六边形
plt.fill(angles,data, alpha=0.25)                      # 填充，alpha是透明度（自己的实践理解）
plt.thetagrids(angles*180/np.pi, radar_labels)         # 放置属性（radar_labels）
plt.figtext(0.52, 0.95, '霍兰德人格分析', ha='center', size=20)         # 放置标题 ，ha是horizontalalignment（水平对齐方式）的缩写
legend = plt.legend(data_labels, loc=(0.94, 0.80), labelspacing=0.1)   # 放置图注（右上角）
plt.setp(legend.get_texts(), fontsize='large')
plt.grid(True)          # 打开坐标网格
plt.savefig('holland_radar.jpg')     # 保存图片
plt.show()              # 显示
````
