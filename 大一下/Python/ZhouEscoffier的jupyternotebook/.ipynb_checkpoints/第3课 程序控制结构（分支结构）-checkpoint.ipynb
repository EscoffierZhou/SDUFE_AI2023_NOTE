## **第3课 程序控制机构（流程图、分支）**
***
1. **程序有三种基本结构组成：顺序结构，循环结构，分支结构**
2. **顺序结构是指程序按照顺序依次进行的运行方式**
3. **分支结构是指（for循环或者while循环）程序按照双分支或多分支的形式运行（注意只能n选1）**
   
   `for i in range(start,stop,sequence)`
5. **循环结构是指（while循环）程序按照循环次数运行程序，其中有关键词continue跳过当前循环，break关键词跳出所有循环**

#### **3.1 程序的基本结构&程序的流程图**

````python
# 微实例 4.1 圆的面积和周长的计算
#micro-ex4.1 circle sourface
import math

r = eval(input("please input r:"))
S = math.pi*r*r
C = 2*math.pi*r
print("面积:{:2f},周长:{:2f}".format(S,C))
````

````python
# 微实例 4.2 实数的绝对值（分支结构）
#micro-ex4.2 abs(r)
r = eval(input("输入实数:"))
if r<0:
    r=-r
print("绝对值{}".format(r))
````

````python
#微实例 4.3 整数累加 计算从1到正整数R的累加和（循环结构）
r = eval(input("输入正整数:"))
i=0
S=0
while (i<=r):
   S = S+i
   i = i+1
print("累加求和{}".format(S))
````

#### **3.3 BMI程序**


````python
#ex5.3CalBMI.py
height, weight = eval(input("请输入身高(米)和体重(公斤)[逗号隔开]: "))
bmi = weight / height**2
print("BMI 数值为：{:.2f}".format(bmi))
who, dom = "", ""
if bmi < 18.5:
    who, dom = "偏瘦", "偏瘦"
elif bmi < 24: 
   who, dom = "正常", "正常"
elif bmi < 25:
    who, dom = "正常", "偏胖"
elif bmi < 28:
    who, dom = "偏胖", "偏胖"
elif bmi < 30:
    who, dom = "偏胖", "肥胖"
else:
    who, dom = "肥胖", "肥胖"
print("BMI 指标为:国际'{0}', 国内'{1}'".format(who,dom))
````

````python
"""
11. (简答题)
编程实现成绩划分等级。用户输入0-100分的成绩，
按照大于等于90分为A，大于等于75分为B，大于等于60分为C，其他成绩为D的规则，输出显示该成绩的等级。
"""
#This is a comment by 周兴 202318140413
score = eval(input("Insert your marks: "))
if 100.0 > score >= 90.0 :
    print("A")
elif 100.0 > score >= 75.0:
    print("B")
elif 100.0 > score >= 60.0:
    print("C")
elif 100.0 > score >= 0:
    print("D")
else:
    print("You put a wrong number out of the range 0-100")
````

````python
"""
12. (简答题)
编写代码实现功能，请在答题框拍照上传代码和输出结果部分。
编程实现身体质量评级自查。请输入自己的身高和体重，代码运行能输出你的bmi值，并自动按照国内BMI的表输出健康等级评价。
"""
#This is a comment by 周兴 202318140413

def calculate(height,weight):
    bmi = weight / (height ** 2)
    return bmi
def get_bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"
number1,number2= eval(input("请输入用','隔开的身高(m)和体重(kg): "))

bmi=calculate(number1,number2)
category=get_bmi_category(bmi)

print("你的bmi是 %.2f,等级是 %s" % (bmi,category))
````
