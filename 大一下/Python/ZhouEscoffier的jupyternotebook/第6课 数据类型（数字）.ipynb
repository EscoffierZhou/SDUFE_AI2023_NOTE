**11. (简答题)编程实现计算三角形的面积。**

**用户输入三角形的三边长，分别用变量a,b,c表示。**

**根据海伦公式（p=（a+b+c)/2），三角形的面积S=sqrt[p(p-a)(p-b)(p-c)]，要求计算结果保留两位有效数字。**

**要求：先输入三边长为3,4,5，验证代码的计算结果是否正确；然后计算三边长是5,6,7时，三角形的面积。**


````python
#Thia is a comment by 202318140413 Zhouxing
import math

str_input = input("请输入三角形的三边长，用逗号隔开")
lengths = str_input.split(',')

a = float(lengths[0])
b = float(lengths[1])
c = float(lengths[2])

p = (a + b + c)/ 2
S = math.sqrt(p * (p-a) * (p-b) * (p-c))
print("三角形的面积是{:.2f}".format(S))

````

**12. (简答题)编程实现判断某个数是否是质数。质数的定义是仅能被1和自己整除的数。**

````python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# 测试
num = int(input("请输入一个整数："))
if is_prime(num):
    print(num, "是质数")
else:
    print(num, "不是质数")
````
