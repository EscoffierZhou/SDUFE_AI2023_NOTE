**1.3 字符串中字符的序号（索引）**
>**Python字符串包括两种序号体系（N:M）**
>
>**正向递增序号**
>
>>**左侧字符序号从0开始，最右侧序号为L-1**
>
>**反向递减序号**
>
>>**右侧字符序号从-1开始，最左侧序号为-L**
>
>**注意的是；M和N任意一项第一可以省略，但是有[]，冒号就不能省略**
>>
>
>**注意： [ M: N] 是前闭后开的区间，是从序号为M的字符开始取，到序号为N-1 的字符；**
>
>>**可以统一理解为[M:N]是获取一段子串，若M=N就是获取一个字符**
>
**1.4 使用索引对字符串进行访问**
>**字符串中字符的访问包括：单个字符的访问（通过索引访问）和多个字符的访问（区间访问，也称为切片访问）**
>
>>**高级表达：**
>>
>>**[M:N:K]表示对字符串多个字符访问，从M开始，到N结束（仅能取N的前一个），切片时步长K（每k个取1个）**
>>
>>**[M: N]中的M和N都可以缺失。若M缺失表示从头开始，若N缺失表示取至结尾，若M和N都缺失表示从头至尾获取字符**
>>
>>**访问单个字符，[ ]中只写一个序号，正向序号或反向序号均可**

#### **2.6.1 format()方式的基本使用**
>**{ <参数序号> ： <格式控制标记> }**
>
**基本思想：将format()方法中逗号分隔的参数按照序号关系替换到模板字符串的槽中**

````python
#This is a comment by hiiragi-ansuke

#String format:

s="I love nireinozomi"
#Warning:30 is the space that ignored
#left(default)         
"{0:30}".format(s)    #'I love nireinozomi            '
"{0:>30}".format(s)   #'            I love nireinozomi'
"{0:*^30}".format(s)  #'******I love nireinozomi******'
"{0:ღ^30}".format(s)  #'ღღღღღღI love nireinozomiღღღღღღ'
"{0:❥^30}".format(s)  #'❥❥❥❥❥❥I love nireinozomi❥❥❥❥❥❥'
````

````python
#This is a comment by hiiragi-ansuke

#decimal format:

x=1234567890
"{0:-^20}".format(x)     #'-----1234567890-----'
"{0:-^20,}".format(x)    #'---1,234,567,890----'

````

````python
#This is a comment by hiiragi-ansuke

#Precise format:

x=1234.567890
"{0:.2f}".format(x)    #'1234.57'
"{0:*^20.3f}".format(x)    #'******1234.568******'

````

**文本进度条**

````python
import time

scale = 10
print("----------ACTIVATED----------")
for i in range (scale+1):
    a,b = '**'*i,'..'*(scale - i)
    c = (i/scale)*100
    print("%{:^3.0f}[{}->{}".format(c,a,b))
    time.sleep(0.1)
print("----------FINISHED----------")
````

````python
# 无限进度条
import time
import random

scale = 10  # 进度条的长度
position = 0  # 进度条的当前位置

print("----------开始----------")
while True:
    # 生成随机的进度百分比
    progress = random.randint(0, 100)

    # 根据进度百分比计算进度条的位置
    position = int((progress / 100) * scale)

    # 根据当前位置计算填充和剩余符号的数量
    filled_symbols = '*' * position
    remaining_symbols = '.' * (scale - position)

    print("%{:^3.0f}[{}->{}".format(progress, filled_symbols, remaining_symbols))  # 打印进度条

    time.sleep(0.1)  # 暂停0.1秒

print("----------完成----------")
````

````python
s = "根据艾瑞数据分析，2019年中国科技企业技术研发投入约为4005亿元，其中人工智能算法研发投入占比为9.3%，超370亿元，且大部分投入来自互联网科技公司。主要Al算法应用领域——计算机视觉、语音识别/语音合成，以及自然语言处理占比分别为22.5%、2.3%和7.1%，三者中计算机视觉相关算法研发投入占比最大，这与视觉相关创业公司数量、产业需求和政策导向呈正相关联系，计算机视觉目前仍是中国最具代表性的Al应用技术。"

#生成式插旗法遍历，然后数旗子
total_chars = len(s)
digit_count = sum(1 for char in s if char.isdigit())
alpha_count = sum(1 for char in s if char.isalpha())
print("总字符:", len(s))

print("数字:", sum(1 for char in s if char.isdigit()))

print("字母:", sum(1 for char in s if char.isalpha()))
print("其他符号:", total_chars - digit_count - alpha_count)
````
