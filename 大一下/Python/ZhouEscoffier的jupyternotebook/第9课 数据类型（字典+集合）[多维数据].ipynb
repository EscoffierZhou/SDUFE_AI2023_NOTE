### **集合相关代码**

````python

# 创建两个集合
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# 打印集合
print("set1:", set1)
print("set2:", set2)

# 检查集合中的元素数量
print("set1中的元素数量:", len(set1))
print("set2中的元素数量:", len(set2))

# 检查元素是否存在于集合中
print("3是否存在于set1中:", 3 in set1)
print("6是否存在于set2中:", 6 in set2)

# 执行交集操作
intersection = set1.intersection(set2)
print("set1和set2的交集:", intersection)

# 执行并集操作
union = set1.union(set2)
print("set1和set2的并集:", union)

# 执行差集操作
difference = set1.difference(set2)
print("set1和set2的差集:", difference)

# 执行对称差集操作
symmetric_difference = set1.symmetric_difference(set2)
print("set1和set2的对称差集:", symmetric_difference)

# 检查一个集合是否是另一个集合的子集或超集
print("set1是否是set2的子集:", set1.issubset(set2))
print("set2是否是set1的超集:", set2.issuperset(set1))
````
