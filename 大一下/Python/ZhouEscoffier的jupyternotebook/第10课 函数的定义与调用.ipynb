**10. (简答题, 30分)**
编写程序，能逆序输出（从大到小的顺序）输出某个值以内所有的质数。

要求 1）定义一个函数 Prime() 能够判断一个数是否是质数；

2）主程序中，完成 获取用户输入的整数，按照从大到小的顺序输出小于该整数的所有质数。
```python
def Prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def main():
    limit = int(input("请输入一个整数: "))
    primes = [i for i in range(2, limit) if Prime(i)]
    primes.sort(reverse=True)
    print("小于该整数的所有质数（从大到小）：")
    for prime in primes:
        print(prime, end=' ')
    print()

main()
```

**10. (简答题, 30分)编写程序，能逆序输出（从大到小的顺序）输出某个值以内所有的质数。**

**1）定义一个函数 Prime( )  能够判断一个数是否是质数；**

**2）主程序中，完成  获取用户输入的整数，按照从大到小的顺序输出小于该整数的所有质数。**

````python
# This is a comment by 周兴 202318140413
import math

def prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2): #质数筛的方法
        if n % i == 0:
            return False
    return True

def main():
    try:
        num = int(input("输入一个整数: "))
    except ValueError:
        print("输入无效，请输入一个整数。")
        return
    for i in range(num, 0, -1):
        if prime(i):
            print(f"{i}是质数")
if __name__ == "__main__":
    main()
````
