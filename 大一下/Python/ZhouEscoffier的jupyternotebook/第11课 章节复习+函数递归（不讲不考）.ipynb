2. **递归函数（数据结构和数学计算中常用）**
>**数学归纳法**
>
>>**证明当n取第一个值n0 时命题成立**
>>
>>**假设当nk 时命题成立，证明当n=nk+1时命题也成立**
>
>**定义方式：**
>
>>**例（初值）：存在一个或多个不需要递归的基例**
>>
>>**链条（关联）：计算过程后面值和前面值的递归关联**


````python
def fact(n): # 阶乘算法的实现
    if n ==	0	:
        return 1
    else :
        return n*fact(n-1)
print(fact(5))
````

## **函数递归案例1 ：科赫雪花分形曲线**


>>>>

````python
#KochDrawV1.py
import turtle
def koch(size, n):
    if n == 0:
        turtle.fd(size) #绘制直线
    else:
        for angle in [0, 60, -120, 60]:  turtle.left(angle) #左转  koch(size/3, n-1)  
#绘制低一级的koch曲线
print(koch(3,3))
````

## **函数递归案例2 ：汉诺塔Hanoi tower问题**

````python
#汉诺塔
def move(n, a, b, c):
    if n == 1:
        print(a, '-->', c)
    else:
        move(n-1, a, c, b)  #把n-1个盘子从 a 移动到 b
        move(1, a, b ,c)    #把1个盘子从 a 移动到 c
        move(n-1, b, a, c)  #把n-1个盘子从 b 移动到 c
                           #一直循环到n==1

move(3, 'A', 'B', 'C')
print(move(6,1,2,3))
````
