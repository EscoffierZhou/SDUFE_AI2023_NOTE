````python
"""
12. ()

bmiBMI
"""
#This is a comment by  202318140413

def calculate(height,weight):
    bmi = weight / (height ** 2)
    return bmi
def get_bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"
number1,number2= eval(input("','(m)(kg): "))

bmi=calculate(number1,number2)
category=get_bmi_category(bmi)

print("bmi %.2f, %s" % (bmi,category))
````
