## **CSV文件**

**数据的维度**
>**一维数据：一组处于对等关系的有序或无序数据，对应数组/集合**
>
>**二维数据：由关联关系数据组成**
>
>**高维数据：通过键值对表示数据构成，采用面向对象的方式组成**

````python
# 创建一个csv文件
txt = [
      ["sec A","sec B","sec C","sec D"]
      ["a1","b1","c1","d1"]
      ["a2","b2","c2","d2"]
      ]
filepath = 'F:/txt.csv'
file = open(filepath,'w',encoding = 'utf-8')
for i in txt:
    file.writelines() #全部写入
file.close()
````

````python
# 读取一个csv文件
filepath = 'F:/txt.csv'
file = open(filepath,'r',encoding = 'utf-8')
txt = file.read() #file.readlines()
file.close()
print(txt)
````

## **实例：文本词频统计**


````python
file = open('hamlet.txt','r')
txt = file.read()
file.close()
#print(len(txt))
txt= txt.lower()
for i in txt:
    if i in"""!:;%^#$@$#%$^&*;,"""
        txt = txt.replace(i,"")
words = txt.split()
print(words[:20])
counts = {}
for i in words:
    counts[i]=counts.get(i,0)+1

myCounts = counts.items()
myCounts = list(myCounts)
print(myCounts[:5])

myCounts.sort(key = lambda x : x[1],reverse = True)
print(myCounts[:10])

for i in range(10):
    print("{}:{}".format(myCounts[i][0],myCounts[i][1]))
````
